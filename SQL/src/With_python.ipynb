```python
import mysql.connector as mc
import pandas as pd
```

```python
# AWS RDS 접속
remote = mc.connect(
    host='localhost',
    user='robot',
    password='1234',
    database='armbase'
)

cur = remote.cursor()
```

```python
# CCTV 테이블 생성
cur.execute('CREATE TABLE cctv(구분 varchar(64), 죄종 varchar(64), 발생검거 varchar(64), 건수 int)')
conn.close()
```

```python
# 데이터 불러오기
df = pd.read_csv('./../data/2020_crime.csv', encoding='euc-kr')
df.head()
```

```python
# 데이터 입력
remote = mc.connect(
    host='AWS RDS',
    user='robot',
    password='1234',
    database='armbase'
)

cur = remote.cursor(buffered=True)

sql = """INSERT INTO cctv VALUES (%s, %s, %s, %s)"""

for i, row in df.iterrows():
    
    cur.execute(sql, tuple(row))
remote.commit()
remote.close()
```

```python
# 데이터 입력
remote = mc.connect(
    host='AWS RDS',
    user='robot',
    password='1234',
    database='armbase'
)

cur = remote.cursor(buffered=True)

sql = """SELECT * FROM cctv"""
cur.execute(sql)
result = cur.fetchall()
remote.commit()
remote.close()
```

```python
df = pd.DataFrame(result)
df.head()
```
