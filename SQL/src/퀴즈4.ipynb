```python
import mysql.connector as mc
import pandas as pd

def execute_query_in(remote, query, gen):
    cur = remote.cursor()
    cur.execute(query)
    if gen:
        remote.commit()
    else:
        result = cur.fetchall()
        cur_columns = [i[0] for i in result]
        print(result)
        print(cur_columns)
        try:
            return pd.DataFrame(result, columns=cur_columns)
        except:
            return pd.DataFrame(result)

def execute_query(query, gen=False):
    remote = mc.connect(
        host='AWS RDS',
        user='robot',
        password='1234',
        database='armbase'
        )
    df = execute_query_in(remote, query, gen)
    remote.close()
    return df
```

```python
query = '''desc movie'''
execute_query(query)
```

```python
query = '''select * from movie
where (date between '2018-01-01' and '2018-12-31')
and (point between 5 and 6)
order by point desc, name asc'''
execute_query(query).iloc[:30]
```

```python
query = '''CREATE TABLE def (
코드 char(8),
명칭 varchar(16),
주소 varchar(64),
전화번호 varchar(16),
영업시간 varchar(16),
재고량 int)'''
execute_query(query)
```

```python
query = '''
select * from def 
where (명칭 like '%부산방향%' or 명칭 like '%서울방향%')
and 주소 like '%고속도로%' 
order by 재고량 desc limit 5
'''
execute_query(query)
```
