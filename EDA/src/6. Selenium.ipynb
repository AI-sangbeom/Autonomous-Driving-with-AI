```python
from selenium import webdriver 
from selenium.webdriver.chrome.service import Service

driver = webdriver.Chrome(service=Service('driver/chromedriver-linux64/chromedriver'))
driver.get('https://www.selenium.dev/')
# driver.close()
```

```python
import time
import random
from tqdm import tqdm
```

- execute_script()
  - 페이지 조작
  - window.scrollTo(x, y)
    - document.body.scrollHeight : 페이지 끝
- execute_script(arguments[0].scrollIntoView(false), link)
  - 특정 위치로 화면 이동

```python
# 페이지 가장 밑으로 이동
driver.execute_script(f'window.scrollTo(0, document.body.scrollHeight)')
```

```python
# 페이지 가장 위로 이동
driver.execute_script(f'window.scrollTo(0, 0)')
```

```python
moves = range(1)

pbar = tqdm(moves,
            total=len(moves),
            ncols=80,
            leave=True)
scaling = 0.7
for i in pbar:
    random_number = (random.random() + random.random())/2
    pbar.set_description(f'{random_number}')
    driver.execute_script(f'window.scrollTo(0, document.body.scrollHeight*{random_number*scaling})')
    time.sleep(random.random()*5)
```

```python
from selenium.webdriver.common.by import By

''' 
find_element는 공백 허용 안함.
공백 대신 '.'으로 표기
'''

element = driver.find_element(By.CLASS_NAME, 'd-flex.justify-content-center.p-5.td-box--100')
print(element.get_attribute('innerHTML'))
```

```python
# class Name 한번 더 인덱싱
child = element.find_element(By.CLASS_NAME, 'selenium')
print(child.text)
```

```python
# 특정 위치로 화면 이동 후 스크린 샷 
driver.execute_script('arguments[0].scrollIntoView(false)', child)
driver.save_screenshot('images/image_01.png')
```

# Menu 다루기

```python
menu = driver.find_element(By.ID, 'main_navbar')

# find_elements : TAG_NAME 활용해서 menu에 해당하는 list 불러오기
li_list = menu.find_elements(By.TAG_NAME, 'li')

for idx, li in enumerate(li_list):
    print(f'====================== {idx} ======================')
    print(li.get_attribute('innerHTML'), end='\n\n')
```

```python
about = li_list[0]
about.click()
link_list = about.find_elements(By.CLASS_NAME, 'dropdown-item')
for link in link_list:
    print(link.text)
```

```python
history = about.find_element(By.LINK_TEXT, 'History')
print(history.text)
history.click()
```

```python
driver.back()
```

# Action Chain : 여러가지 연속으로

```python
from selenium.webdriver import ActionChains
import time

actions = ActionChains(driver)
actions.click(about)
actions.click(history)
actions.perform()
```
