## Polium 기초

```python
import folium
```

```python
m = folium.Map(location=[45.5236, -122.6750])
m
```

```python
Tiles_Builtin = ['OpenStreetMap', 'Cartodb Positron', 'Cartodb dar_matter']
folium.Map(location=[37.4778772, 126.8806482], tiles=Tiles_Builtin[0], zoom_start=13)
```

```python
folium.Map(location=[37.4778772, 126.8806482], tiles=Tiles_Builtin[0], zoom_start=18)
```

```python
MyHome = [37.4778772, 126.8806482]
MyNeighbor = [37.47791, 126.8791]

my_map = folium.Map(location=MyHome, tiles=Tiles_Builtin[0], zoom_start=18)

folium.Marker(MyHome, popup='<i>My Sweet Home</i>').add_to(my_map)
folium.Marker(MyNeighbor, popup='<i>My MyNeighbor</i>').add_to(my_map)

my_map
```

```python
MyHome = [37.4778772, 126.8806482]
MyNeighbor = [37.47791, 126.8791]
MyFriend = [37.477822, 126.88062]

m = folium.Map(location=MyHome, tiles=Tiles_Builtin[0], zoom_start=18)

folium.Marker(MyHome, 
              popup='My Sweet Home',
              icon=folium.Icon(color='green')
              ).add_to(m)
folium.Marker(MyNeighbor, 
              popup='My MyNeighbor',
              icon=folium.Icon(color='cloud')
              ).add_to(m)
folium.Marker(MyFriend, 
              popup='My MyFriend',
              icon=folium.Icon(color='red', icon='info-sign')
              ).add_to(m)
m
```

```python
MyHome = [37.4778772, 126.8806482]
MyNeighbor = [37.47791, 126.8791]
MyFriend = [37.477822, 126.88062]

m = folium.Map(location=MyHome, tiles=Tiles_Builtin[0], zoom_start=18)

folium.Circle(
    radius = 100,
    location = MyHome,
    popup='My Sweet Home',
    color='crimson',
    fill=False
).add_to(m)
folium.CircleMarker(
    radius=50,
    location=MyNeighbor, 
    popup='My MyNeighbor',
    color='#3186cc',
    fill=True,
    fill_color='#3186cc'
).add_to(m)
m
```

```python
m = folium.Map(location=MyHome, tiles=Tiles_Builtin[0], zoom_start=18)
m.add_child(folium.LatLngPopup())
m
```
