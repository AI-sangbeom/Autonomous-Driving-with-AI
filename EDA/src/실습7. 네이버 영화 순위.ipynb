## 네이버 영화 순위

```python
import urllib
import urllib.request
from bs4 import BeautifulSoup
from urllib.request import urlopen
from urllib.request import Request

###############################################
# Referer 및 기타 헤더 설정을 통해 403에러 회피
headers = {
    "User-Agent": "Mozilla/5.0...",
    "Referer": "http://naver.com/previous-page"
}
###############################################

url = 'https://search.naver.com/search.naver?where=nexearch&sm=tab_etc&qvt=0&query={search_word}'
req = Request(url.format(search_word=urllib.parse.quote('박스오피스')), headers=headers)

res = urlopen(req)
# res = open('../data/Naver_html.html').read()

soup = BeautifulSoup(res, 'html.parser')
print(soup.prettify())
```

```python
import time
import numpy as np
def get_title(url, headers):
    req = Request(url, headers=headers)
    res = urlopen(req)
    soup = BeautifulSoup(res, 'html.parser')
    return soup.find('strong', class_='_text').string
```

```python
rating = soup.find(class_='_panel_popular _tab_content')
movie_sorted_by_pop = {}

prefix = 'https://search.naver.com/search.naver'
for page in rating.find_all('li'):
    href = page.find('a')['href']
    link = prefix + href
    title = get_title(link, headers)
    pop = page.find('span', class_='sub_text').string
    ranking = page.find('span', class_ = 'this_text').string
    movie_sorted_by_pop[ranking] = {
        'title' : title,
        'pop' : pop,
        'ranking' : ranking,
        'link' : prefix + href
    }
```

```python
rating = soup.find(class_='_panel_rating _tab_content')
movie_sorted_by_rating = {}

prefix = 'https://search.naver.com/search.naver'

for page in rating.find_all('li'):
    href = page.find('a')['href']
    link = prefix + href
    title = get_title(link, headers)
    rating = page.find('div', class_='title_box').find('span').get_text()
    ranking = page.find('span', class_ = 'this_text').string
    movie_sorted_by_rating[ranking] = {
        'title' : title,
        'rating' : rating,
        'ranking' : ranking,
        'link' : link
    }
    time.sleep(0.5)
```

```python
import pandas as pd 

df = pd.DataFrame(movie_sorted_by_pop).transpose()
df.set_index(keys='ranking')
```

```python
df = pd.DataFrame(movie_sorted_by_rating).transpose()
df.set_index(keys='ranking')
```
