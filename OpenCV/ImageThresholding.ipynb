```python
import cv2 as cv
import matplotlib.pyplot as plt
img = cv.imread('./../image/rainbow.jpg')
plt.imshow(img)
```

```python
img = cv.imread('./../image/rainbow.jpg', cv.IMREAD_GRAYSCALE)
plt.imshow(img, 'gray')
```

```python
ret, th1 = cv.threshold(img, 127, 255, cv.THRESH_BINARY)
plt.subplot(1, 2, 1)
plt.imshow(th1, 'gray')
plt.subplot(1, 2, 2)
plt.imshow(img, 'gray')
plt.show()
```

```python
ret, th1 = cv.threshold(img, 127, 255, cv.THRESH_TRUNC)
plt.subplot(1, 2, 1)
plt.imshow(th1, 'gray')
plt.subplot(1, 2, 2)
plt.imshow(img, 'gray')
plt.show()
```

```python
ret, th1 = cv.threshold(img, 127, 255, cv.THRESH_TOZERO_INV)
plt.subplot(1, 2, 1)
plt.imshow(th1, 'gray')
plt.subplot(1, 2, 2)
plt.imshow(img, 'gray')
plt.show()
```

```python
img = cv.imread('./../image/crossword.jpg', 0)
plt.imshow(img, 'gray')
```

```python
def show_pic(img):
    fig = plt.figure(figsize=(15, 15))
    ax = fig.add_subplot(111)
    ax.imshow(img, 'gray')
```

```python
ret, th1 = cv.threshold(img, 127, 255, cv.THRESH_BINARY)
show_pic(th1)
```

```python
ret, th2 = cv.threshold(img, 180, 255, cv.THRESH_BINARY)
show_pic(th2)
```

```python
th2 = cv.adaptiveThreshold(img, 255, cv.ADAPTIVE_THRESH_MEAN_C, cv.THRESH_BINARY, 11, 8)
show_pic(th2)
```

```python
blended = cv.addWeighted(th1, 0.6, th2, 0.4, 0)
show_pic(blended)
```
