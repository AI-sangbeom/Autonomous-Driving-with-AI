```python
import os
import sys
import numpy as np
import cv2 as cv
import matplotlib.pyplot as plt 
```

```python
path = './../image'
src1 = cv.imread(os.path.join(path, 'lenna.bmp'))
src2 = cv.imread(os.path.join(path, 'sky.bmp'))
resize_src2 = cv.resize(src2, (src1.shape[1], src1.shape[0]))
dst = cv.addWeighted(src1, 0.5, resize_src2, 0.5, 0)
show_dst = cv.cvtColor(dst, cv.COLOR_BGR2RGB)
plt.imshow(show_dst)
```

```python
src1 = cv.imread(os.path.join(path, 'lenna.bmp'), cv.IMREAD_GRAYSCALE)
src2 = cv.imread(os.path.join(path, 'circle.jpg'), cv.IMREAD_GRAYSCALE)
src2 = cv.resize(src2, (src1.shape[1], src1.shape[0]))
dst = cv.subtract(src1, src2)

plt.imshow(dst, cmap='gray')
```

```python
img1 = cv.imread(os.path.join(path, 'road1.jpg'), cv.IMREAD_GRAYSCALE)
img2 = cv.imread(os.path.join(path, 'road2.jpg'), cv.IMREAD_GRAYSCALE)

dst = cv.absdiff(img1, img2)
fig = plt.figure(figsize=(10, 7))

fig.add_subplot(1, 3, 1)
plt.imshow(img1, cmap='gray')

fig.add_subplot(1, 3, 2)
plt.imshow(img2, cmap='gray')

fig.add_subplot(1, 3, 3)
plt.imshow(dst, cmap='gray')   
```

```python
import numpy as np
import cv2 as cv
```

```python
src = cv.imread('./../image/lenna.bmp', cv.IMREAD_GRAYSCALE)

if src is None:
    print('Image load failed!')
    sys.exit()

alpha = 1.0
dst = np.clip((1 + alpha) * src - 128 * alpha, 0, 255).astype(np.uint8)

cv.imshow('src', src)
cv.imshow('dst', dst)
cv.waitKey()
cv.destroyAllWindows()
```

```python
src = cv.imread('./../image/lenna.bmp', cv.IMREAD_GRAYSCALE)

if src is None:
    print('Image load failed!')
    sys.exit()

dst = cv.normalize(src, None, 0, 255, cv.NORM_MINMAX)

cv.imshow('src', src)
cv.imshow('dst', dst)
cv.waitKey()
cv.destroyAllWindows()
```

```python
import sys
import numpy as np
import cv2
src = cv2.imread('./../image/horse.jpg', cv2.IMREAD_GRAYSCALE)
if src is None:
    print('Image load failed!')
    sys.exit()
dst = cv2.equalizeHist(src)
cv2.imshow('src', src)
cv2.imshow('dst', dst)
cv2.waitKey()
cv2.destroyAllWindows()


# 컬러 영상의 히스토그램 평활화
src = cv2.imread('./../image/field.bmp')
if src is None:
    print('Image load failed!')
    sys.exit()
src_ycrcb = cv2.cvtColor(src, cv2.COLOR_BGR2YCrCb)
ycrcb_planes = list(cv2.split(src_ycrcb))

# 밝기 성분에 대해서만 히스토그램 평활화 수행
ycrcb_planes[0] = cv2.equalizeHist(ycrcb_planes[0])
dst_ycrcb = cv2.merge(ycrcb_planes)
dst = cv2.cvtColor(dst_ycrcb, cv2.COLOR_YCrCb2BGR)
cv2.imshow('src', src)
cv2.imshow('dst', dst)
cv2.waitKey()
cv2.destroyAllWindows()
```
